```python
import numpy as np
import pandas as pd

df = pd.read_json('contacts.json')
df
```

```python
orderID = df[['Id','OrderId']]
orderID['OrderId'].replace('', np.nan, inplace=True)
orderID.dropna(subset=['OrderId'], inplace=True)
orderID
```

```python
duplicateOrder = orderID[orderID.duplicated('OrderId')] 
orderDict = duplicateOrder.index.groupby(duplicateOrder['OrderId'])
finOrder = pd.DataFrame.from_dict(orderDict, orient='index')
finOrder['OrderId'] = finOrder.index
finOrder
```

```python
email = df[['Id','Email']]
email['Email'].replace('', np.nan, inplace=True)
email.dropna(subset=['Email'], inplace=True)

duplicateEmail = email[email.duplicated('Email')] 
emailDict = duplicateEmail.index.groupby(duplicateEmail['Email'])
finEmail = pd.DataFrame.from_dict(emailDict, orient='index')
finEmail['Email'] = finEmail.index
finEmail
```

```python
phone = df[['Id','Phone']]
phone['Phone'].replace('', np.nan, inplace=True)
phone.dropna(subset=['Phone'], inplace=True)

duplicatePhone = phone[phone.duplicated('Phone')] 
phoneDict = duplicatePhone.index.groupby(duplicatePhone['Phone'])
finPhone = pd.DataFrame.from_dict(phoneDict, orient='index')
finPhone['Phone'] = finPhone.index
finPhone
```

```python
dfO = pd.merge(df, finOrder, how='left', on='OrderId')
dfE = pd.merge(df, finEmail, how='left', on='Email')
dfP = pd.merge(df, finPhone, how='left', on='Phone')

dfOE = pd.merge(dfO, dfE, how='outer', on='Id')
dfALL = pd.merge(dfOE, dfP, how='outer', on='Id')
dfALL
```

```python

defALL =  dfALL.replace(np.nan, '', regex=True)

defALL.loc[dfALL['Id'], 'output'] = defALL['Id'].astype(str) + '-' + defALL[0].astype(str) + '-' + defALL[1].astype(str)+ '-' + defALL[2].astype(str) + '-' + defALL['0_x'].astype(str) + '-' + defALL['1_x'].astype(str)+ '-' + defALL['2_x'].astype(str)+ '-' + defALL['0_y'].astype(str) + '-' + defALL['1_y'].astype(str) + '-' + defALL['2_y'].astype(str)
defALL
```

```python
out = defALL[['Id', 'output']]
out = out.rename(columns={"Id" : "ticket_id"})
out = out.rename(columns={"Id" : "ticket_id"})
out
```
