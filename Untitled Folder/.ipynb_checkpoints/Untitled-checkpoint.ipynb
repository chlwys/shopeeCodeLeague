```python
import numpy as np
import pandas as pd

df = pd.read_json('contacts.json')
df
```

```python
orderID = df[['Id','OrderId']]
orderID['OrderId'].replace('', np.nan, inplace=True)
orderID.dropna(subset=['OrderId'], inplace=True)
orderID
```

```python
duplicateOrder = orderID[orderID.duplicated('OrderId')] 
duplicateOrder
```

```python
duplicateOrder.index.groupby(duplicateOrder['OrderId'])
```

```python
orderDict = {k:list(v) for k,v in duplicateOrder.index.groupby(duplicateOrder['OrderId']).items()}
for order, ids in orderDict.items():
    if (len(ids) < 3):
        for x in range(0,3-len(ids)):
            ids.append(0)
        

orderDict
#pd.DataFrame.from_dict(orderDict)
```
